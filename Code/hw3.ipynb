```python
# Question 1

## validating personal and domain
# letters, digits, and special are lists that consist of acceptable characters

letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 
           'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 
           'u', 'v', 'w', 'x', 'y', 'z']
    
digits = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']

char_domain = [".", "-"]

char_personal = ["!", "#","$", "%", "&", "'", "*", "+", "/", "=", "?", "^", "_", "{", "|", "}", "~", ".", "-"]
    
    
def check_email_validity(mail_str):
    
    # Reference for email validity: https://en.wikipedia.org/wiki/Email_address#Syntax

    
    # Counter based on '@' counts
    counter = 0
    for _ in mail_str:
        if _ == '@':
            counter += 1

    if counter !=1:
        return None
    else:
        pass
    
    
    # consecutive periods
    if ".." in mail_str:
        return None
    else:
        pass
        
    mail_str1 = mail_str.replace("<", "")
    mail_str2 = mail_str1.replace(">", "")
    
    # Splitting the email into personal and domain
    personal = mail_str2.split("@")[0].lower()
    domain = mail_str2.split("@")[1].lower()
    

    # validating based on length
    if len(personal)>64 or len(personal)<1:
        return None
    elif len(domain)>253 or len(domain)<1:
        return None
    else:
        pass

    
    # Checking personal for period (".") conditions
    if personal[0] == "." or personal[len(personal) - 1] == ".":
        return None
    else:
        pass
    
    
    # Checking domain for period (".") conditions
    if domain[0] == "." or domain[len(domain)-1] == "." or domain[0] == "-" or domain[len(domain)-1] == "-":
        return None
    else:
        pass
    
    
    # Looping through the entire personal string for validation:
    for i in range(len(personal) - 1):
        if (not personal[i] in letters and not personal[i] in digits and not personal[i] in char_personal):
            return None
        else:
            pass
        
        
    #Looping through domain for validation
    for i in range(len(domain) - 1):
        if (not domain[i] in letters and not domain[i] in char_domain and not domain[i] in digits):
            return None
        else:
            pass
        
    return (mail_str2)
```

```python
# Question 2
def process_newsgroup_file(file_path, word_dict):
    
    # Using keyword 'From:' to identify the sender's email
    
    word_email = 'From:'
    
    # Using keywork 'Lines:' to use as a separator between header and content based on personal observation.
    # All files have the above keyword, and it always occurs after 'From:' in the header
    
    word_lines = 'Lines:' 
    res = None
    counter = 0
    indice = 1
    
    # Processing the email validity
    with open (file_path, 'r', encoding = 'latin-1') as file:
        lines = file.readlines()
        
        # Finding the first blank line to separate header from content 
        for header_line in lines:
            if header_line != "\n":
                indice += 1
            else:
                break
            
            
        for mail_line in lines:
        
            # Email validity
            if mail_line.find(word_email) != -1:
                #print(word, 'string exists in file')
                #print('Line Number:', lines.index(line))
                #print('Line:', line)
                email_line = mail_line.split(" ")
                
                for i in email_line:
                    if "@" in i:
                        email_id = check_email_validity(i)
                    else:
                        pass
                            

                
                print(email_id)
                continue
            else:
                pass
    
            # Finding empty lines. First one indicates separation of header from content
            #if line == '\n':
             #   indice = lines.index(line)
            #else:
             #   pass
        
        # Scan through the lines after the first empty line to count quotes
        for line in lines:
            if lines.index(line) < indice:
                pass
            else:
                if line[0] == '>':
                    counter += 1
                else:
                    pass
        
                if "u'\xa0'" in line: 
                    line.replace(u'\xa0', ' ')
                else:
                    pass
        
                for i in line:
                    if not i in letters or not i in digits:
                        new_string = line.replace(i, '')
                    else:
                        pass
        
                n_line = new_string
    
                words_list = n_line.split()
        
                for j in words_list:
                    if j in word_dict.keys():
                        word_dict[j] += 1
                    else:
                        word_dict[j] = 1
        
    return(email_id, counter)
```

```python
# Question 3

import glob
import pandas as pd

word_list = {}
df = pd.DataFrame(columns = ["file", "email", "quotes"])
dfw = pd.DataFrame(columns = ["word", "count"])

files = glob.glob("C:/Users/arjun/DATA515/hw3-Arjun-SC31/20_newsgroups/sci.crypt/*")  # adjust filepath as needed
filename = []
emails = []
quote_count = []

for f in files:
    filename.append(f.replace("C:/Users/arjun/DATA515/hw3-Arjun-SC31/20_newsgroups/sci.crypt\\", ""))
    mail, words = process_newsgroup_file(f, word_list)
    emails.append(mail)
    quote_count.append(words)
    
    
df["file"] = filename
df["email"] = emails
df["quotes"] = quote_count

dfw["word"] = word_list.keys()
dfw["count"] = word_list.values()

df.to_csv("email_quotes2.csv", index = False)
dfw.to_csv("word_count2.csv", index = False)
```
